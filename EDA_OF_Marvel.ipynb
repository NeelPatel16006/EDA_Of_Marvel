```python
# good super characters marvel

good=marvel[marvel['Alignment']=='good']
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power=good[good['Power']==100]
```

```python
good_max_power
```

```python
good_max_power.shape
```

```python
good_max_power=good_max_power.sort_values(by=['Total'],ascending=False).head()
```

```python
# ploting max power good charactes using bar
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[:5],list(good_max_power['Total'])[:5],color=['pink','yellow','blue','orange','olive'])
plt.xticks(rotation='vertical')
plt.show()
```
